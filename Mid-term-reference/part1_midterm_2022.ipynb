### You are ready to complete class ```Ratings```. 

````python
class Ratings: 
    
    def __init__(self):
        self._size = 5
        self._ratings = []
        
    ### TODO 1.1 -----------------------------------
    def _is_valid_rating(self, r):
        if r > 0 and r <= 5:
            return True
        return False
    
    ### TODO 1.2 -----------------------------------
    @property
    def rating(self):
        n = len(self._ratings)
        if len(self._ratings) == 0:
            return 0
        total = sum(self._ratings)
        return total / n

        
    ### TODO 1.3 -----------------------------------
    @rating.setter
    def rating(self, value):
        if not self._is_valid_rating(value):
            return
        if len(self._ratings) == self._size:
            self._ratings.pop(0)
            self._ratings.append(value)
        if len(self._ratings) < self._size:
            self._ratings.append(value)
        
            
    def __str__(self):
        return str(self._ratings)
````

### You are ready to complete class ```Restaurant```. 

````python
class Restaurant: 
    
    def __init__(self, name, food): 
        self._dd = {'name': 'default', 'food': 'default'}
        
        ### TODO 1.4 --------------------------
        if isinstance(name, str) and isinstance(food, str):
            self._dd['name'] = name
            self._dd['food'] = food
            
        self._ratings = Ratings()
        

    ### TODO 1.5 --------------------------
    @property
    def name(self):
        return self._dd['name']
    
    ### TODO 1.6 --------------------------
    @property
    def food(self):
        return self._dd['food']
    
    ### TODO 1.7 --------------------------
    @property
    def rating(self):
        return self._ratings.rating
    
    @rating.setter
    def rating(self, value):
        self._ratings.rating = value
        
    def __str__(self):
        return self.name + '*' + self.food + '*' + str(self.rating)
````

> **Algorithm 1.1** 
>
> 1. ```A``` is an array to be sorted. 
> 2. ```n``` is the number of elements in ```A```
> 3. ```max_iters``` is the integer value of ```n/2``` rounded down e.g. if the result is 3 or 3.5, 3 is used. 
> 3. ```end``` is ```n-1```
> 3. for iteration ```i``` from ```1``` to ```max_iters```, both values inclusive
>
>    a. for elements ```j``` incremented by ```1``` from index ```i``` to ```end```, both values inclusive
>
>        i. if A[j] < A[j-1], swap, else, do nothing
>
>    b. for elements ```j``` incremented by ```-1``` from index ```end``` to ```i```, both values inclusive
>
>        i. if A[j] < A[j-1], swap, else, do nothing
>
>    c. reduce ```end``` by 1

````python
class Feature:
    
    def __init__(self):
        self._restaurants = []
        
    def add_restaurant(self, r):
        self._restaurants.append(r)
        
    ### TODO 1.8 ---------------------------
    def sort_restaurants(self, category):
        if category == 'food':
            self._sort(lambda f: f.food)
        if category == 'name':
            self._sort(lambda f: f.name)
        if category == 'rating':
            self._sort(lambda f: f.rating)
        
    ### TODO 1.9 ---------------------------
    def _sort(self, f):
        a = self._restaurants
        n = len(self._restaurants)
        max_iters = n // 2
        end = n - 1
        for i in range(1, max_iters + 1):
            for j in range(1, end + 1):
                if f(a[j]) < f(a[j - 1]):
                    a[j], a[j - 1] = a[j - 1], a[j]
            for j in range(end, i + 1, -1):
                if f(a[j]) < f(a[j-1]):
                    a[j], a[j - 1] = a[j - 1], a[j]
            end -= 1
                    
    def __str__(self):
        out = ''
        for s in self._restaurants:
            out = out + str(s) + '\n'
            
        return out 
                    
````

````python
a = Restaurant('Bremen', 'Germany')
a.rating = 3

b = Restaurant('Anglesey', 'Wales')
b.rating = 4

c = Restaurant('Danzig', 'Poland')
c.rating = 5

f = Feature()
f.add_restaurant(a)
f.add_restaurant(b)
f.add_restaurant(c)
f.sort_restaurants('rating')
print('-----')
print(f)
print('-----')
expected = """Bremen*Germany*3.0
Anglesey*Wales*4.0
Danzig*Poland*5.0
"""
assert str(f) == expected 
````
