# Task 3: Shortest Path in Weighted Graph

You can get the shortest path for non-weighted graph using a Breadth-First Search algorithm. For shortest path on a weighted graph, you will need to modify the BFS algorithm to take into account the weight.

In general, the modifications look something like the following:
1. Initialize all the distances as a very large number except the starting Vertex. Set the distance of the starting vertex to be zero.
1. Put all the vertices into a min-priority Queue
1. As long as the queue is not empty
    1. Extract the min priority element from the queue
    1. For every neighbour of that vertex
        1. Relax the distance of the neighbouring Vertex

## Note 1: Min-Priority Queue

Instead of using a normal Queue, we now need to use a min-priority Queue. You can implement a min-priority Queue using a min-heap. Given a list, you can heapify the list into a min-heap as follows.

```python
import heapq

my_list = [3, 4, 2, 1]
heapq.heapify(my_list)
```

The first element will be the minimum in the list and so you can pop out the minimum in the list using the following:
```python
out = my_list.pop(0)
```

````python
import heapq

my_list = [3, 4, 2, 1]
heapq.heapify(my_list)
print(my_list)
````

You can use this for a list of tuple. By default, it will heapify using the first element of the tuple.


````python
import heapq

my_list = [(3, 'a'), (4, 'b'), (2, 'c'), (1, 'd')]
heapq.heapify(my_list)
print(my_list)
````


## Calculate Distance Function

You have to implement the `calc_dist(graph, x)` function which outputs a list which contains the shortest distance from `node x` to every other nodes in the graph.

Eg.
- input: `calc_dist(graph, 1)`

- output: `out = [1, 0, 1, 1, 1, 2]`

The `calc_dist(graph, 1)` outputs the list `[1, 0, 1, 1, 1, 2]` in which each index of `out` represents the nodes of the graph, i.e. `out[0]` is the distance from node 1 to node `0`,  `out[5]` is the distance from node 1 to node 5. The second value in the index is 0 because index 1 in `out` is for node 1 itself and so it has zero distance. 

````python
import heapq

graph = { 
    0: [(1, 3), (2, 5)],
    1: [(0, 3), (2, 3), (3, 3), (4, 9)],
    2: [(0, 5), (1, 3), (3, 5), (4, 4)],
    3: [(1, 3), (2, 5), (4, 5), (5, 9)],
    4: [(1, 9), (2, 4), (3, 5), (5, 3)],
    5: [(3, 9), (4, 3)]
}

def calc_dist(graph, startv):
    dist = [float('inf')] * len(graph)
    dist[startv] = 0
    visited = [False] * len(graph)
    queue = [(0, startv)]

    while queue:
        curr_dist, node = heapq.heappop(queue)

        if visited[node]:
            continue
        visited[node] = True

        for neighbor, weight in graph[node]:
            if dist[node] + weight < dist[neighbor]:
                dist[neighbor] = dist[node] + weight
                heapq.heappush(queue, (dist[neighbor], neighbor))

    return dist



    # Initialize all the distances as a very large number except the starting Vertex. Set the distance of the starting vertex to be zero.
    # Put all the vertices into a min-priority Queue
    # As long as the queue is not empty
    #     Extract the min priority element from the queue
    #     For every neighbour of that vertex
    #         Relax the distance of the neighbouring Vertex
print(calc_dist(graph,0))
````
